```python
#1 Tạo Dataframe như dữ liệu đã cho trên
import pandas as pd

data = {
    "Name": ["Mr.Harris", "Mrs.Bradley", "Miss. Laina", "Mrs.Heath", "Mr.William", "Mr.James", "Mr.Timothy", "Mrs.Oscar", "Mrs.Nicholas", "Miss.Marguerite", "Mr.William", "Miss. Hulda", "Mr.Lawrence", "Mr.Wilam"],
    "Height": [1.80, 1.61, 1.78, 1.74, 1.55, 1.9, 1.61, 1.40, 1.92, 1.77, 1.55, 1.85, 1.64, 1.77],
    "Weight": [90, 87, 64, 70, 100, 106, 82, 55, 77, 60, 100, 109, 80, 66],
    "AGE": [32,45,22,52,48,21,28,69,58,21,48,33,25,70]
}
df = pd.DataFrame(data)
print(df)
```

```python
#Hiển thị thông tin Dataframe vừa tạo
print(df.info())
```

```python
#Tính giá trị trung bình của cột <AGE>
mean_AGE = df['AGE'].mean()
print(mean_AGE)
```

```python
#Nhóm dữ liệu cột <Height> và tính trung bình cột <Weight>
grouped_df = df.groupby('Height')['Weight'].mean()
print(grouped_df)
```

```python
#Sắp xếp Dataframe theo cột <Weight> tăng dần
df_sorted = df.sort_values('Weight')
print(df_sorted)
```

```python
#Vẽ biểu đồ cột cho <AGE>
import matplotlib.pyplot as plt
plt.bar(df.index, df['AGE'])
plt.xlabel('INDEX')
plt.ylabel('AGE')
plt.title('Biểu đồ cột cho AGE')
plt.show()
```

```python
#Vẽ biểu đồ phân tán cho <AGE> và <Height>
plt.scatter(df['AGE'], df['Height'])
plt.xlabel('AGE')
plt.ylabel('Height')
plt.title('Biểu đồ phân tán AGE và Height')
plt.show()
```

```python
#Vẽ biểu đồ đường cho cột <Weight>
plt.plot(df['Weight'])
plt.xlabel('Index')
plt.ylabel('Weight')
plt.title('Biểu đồ đường cho cột Weight')
plt.show()
```

```python
#Vẽ biểu đồ tròn cho cột <Height>
plt.pie(df['Height'].value_counts(), labels=df['Height'].unique())
plt.title('Biểu đồ tròn cho cột Height')
plt.show()
```

```python
#Vẽ biểu đồ phân tán cho cột <Weight>,<AGE>
plt.scatter(df['Weight'], df['AGE'])
plt.xlabel('Weigjt')
plt.ylabel('AGE')
plt.title('Biểu đồ phân tán Weight và AGE')
plt.show()
```

```python
#Kiểm tra có giá trị NaN nào trong DataFrame không
print(df.insull().sum())
```

```python
#Tìm các giá trị trùng lặp trong DataFrame dựa trên cột' Name' và loại bỏ các hàng trung lặp, giữ lại hàng đầu tiên
df_no_duplicates = df.drop_duplicates(subset='Name', keep='first')
print(df_no_duplicates)
```

```python
#Thay thế các giá trị của cột <AGE> lớn hơn 40 bằng giá trị trung bình của cột đó
mean_AGE = df['AGE'].mean()
df.loc[df['AGE'] > 40, 'AGE'] = mean_AGE
print("DataFrame sau khi thay thế:")
print(df)
```

```python
#Tạo một cột mới trong DataFrame với tên <cs_BMI>, tính chỉ số BIM mỗi người 
df['cs_BMI'] = df['Weight'] / (df['Height'] / 100)**2
print("DataFrame sau khi thêm một cột cs_BMI")
print(df)
```

```python
#Tạo cột với tên <Phan_Loai> với các mức '0','1','2','3' tương ứng với mức chỉ số BMI trên
def phan_loai_bmi(bmi):
    if bmi < 18.5:
        return 0
    elif 18.5 <= bmi < 23:
        return 1 
    elif 23 <= bmi < 25:
        return 2 
    else:
        return 3
df['Phan_Loai'] = df['cs_BMI'].apply(phan_loai_bmi)
print("DataFrame sau khi thêm cột Phan_Loai:")
print(df)
```

```python
#Chuẩn hóa giá trị của cột <cs_BMI> chuẩn hóa giá trị trong đoạn từ 0 tới 1
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df['cs_BMI_normalized'] = scaler.fit_transform(df[['cs_BMI']])
print("DataFrame sau khi chuẩn hóa cs_BMI:")
print(df)
```

```python
#Tạo cột với tên <GT>, Sau khi thực hiện, xóa các từ Mr, Mrs, Miss nếu trong cột <Name> còn các từ đó.
def extract_gender(name):
    if " Mr " in name:
        return "male"
    elif "Mrs" in name or "Miss" in name:
        return "female"
    else:
        return "unknown"
df['GT'] = df['Name'].apply(extract_gender)
df['Name'] = df['Name'].str.replace(r'(Mr\.|Mrs\.|Miss\.)\s*','',regex=True)
print("DataFrame sau khi tạo cột GT và xóa các từ")
print(df)
```

```python
#Vẽ biểu đồ tròn cho <GT>
gender_counts = df['GT'].value_counts()
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%')
plt.title("Phân phối giới tính")
plt.show()
```

```python
#Trực quan hóa dữ liệu người nữ trên <Phan_Loai>
female_data = df[df['GT'] == 'female']
phan_loai_counts = female_data['Phan_Loai'].value_counts().sort_index()
plt.bar(phan_loai_counts.index, phan_loai_counts.values)
plt.xlabel("Phân loại BMI (0: thiếu cân, 1: bình thường, 2: thừa cân, 3: béo phì)")
plt.ylabel("Số lượng người nữ")
plt.title("Phân loại BMI của người nữ")
plt.xticks(phan_loai_counts.index)
plt.show()
```
