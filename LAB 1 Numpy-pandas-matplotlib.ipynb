1.1 TẠO MẢNG VỚI NUMPY

```python
import numpy as np
# TẠO MẢNG 1 CHIỀU
mang_1_chieu = np.array([1,3,5,7,9])
print(mang_1_chieu)
# TẠO MẢNG HAI CHIỀU
mang_2_chieu = np.array([[1,3,5,7,9],[2,4,6,8,10]])
print(mang_2_chieu)
# TẠO MẢNG NGẪN NHIÊN
mang_ngau_nhien = np.random.rand(3,3)
print(mang_ngau_nhien)
```

Các thao tác cơ bản của numpy


```python
mang_1_chieu = np.array([1,3,5,7,9])

#Cộng trừ nhân chia với một số
sum = mang_1_chieu + 10
multiple = mang_1_chieu * 5

print(sum)
print(multiple)

#Tính tổng, trung bình, lớn nhất, nhỏ nhất
print(mang_1_chieu.sum())
print(mang_1_chieu.mean())
print(mang_1_chieu.max())
print(mang_1_chieu.min())

#Lấy phân tử bất kỳ trong mảng

print(mang_1_chieu[3])
```

Pandas

```python
import pandas as pd

data = {
    "Ten": ["Anh", "Đạt", "Trâm", "Linh"],
    "Tuổi": [30, 50, 70, 100],
    "Điểm": [55,85,100,95]
}
df = pd.DataFrame(data)
print(df)
```

```python
# In ra cột tuổi
print(df["Tuổi"])
# In ra hàng đầu tiên
print(df.iloc[0])
# In ra dữ liệu có điềm lớn hơn 80 
print(df[df["Điểm"]>80])
# Tính trung bình điểm lớp học
print(df["Điểm"].mean())
```

 ### Matplotlib

```python
import matplotlib.pyplot as plt
```

```python
import matplotlib.pyplot as plt

# Vẽ biểu đồ đường
x =[1,3,5,7,8]
y =[3,9,87,5,4]

plt.plot(x, y, label = " Biểu đồ đường", color = "red", marker = "o" )
plt.title("Biểu đồ đường")
plt.xlabel("Trục X")
plt.ylabel("Trục Y")
plt.legend()                   
plt.show()
```

```python
# Biểu đồ cột

Ten = ["Ly", "Mai", "Xuan", "Bich", "An", "Khang", "Huy"]
Diem = [5,7,3,6,1,0,10]

plt.bar(Ten,Diem, color="yellow")
plt.title("Biểu đồ cột")
plt.xlabel("Tên")
plt.ylabel("Điểm")
plt.show()
```

```python
# Biểu đồ phân tán

x = np.random.rand(100)
y = np.random.rand(100)
size = np.random.rand(100) * 100

plt.scatter(x,y, size, alpha = 0.5, c='pink')
plt.title("Biểu đồ phân tán")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```
